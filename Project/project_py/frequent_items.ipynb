```python
# Read the Data

# Todo, slice years

def read_data (csv_file, number_rows_to_read=None, slice_year=None, slice_year_column=None, columns_with_dates=None, skip=None):
    #dateparser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
    dateparser = lambda x: pd.to_datetime(x, format='%Y-%m-%d', coerce=True)

    df_data = pd.read_csv(csv_file,
                     delimiter= "\t", 
                     encoding= "utf-8",
                     nrows= number_rows_to_read, 
                     parse_dates = columns_with_dates,
                     infer_datetime_format = True,
                     date_parser=dateparser,
                     skiprows = skip,
                     low_memory=False
                    )
    if slice_year != None:
        df_data = df_data[df_data[slice_year_column].dt.year == slice_year]
    return df_data
```

```python
# Constants

file_name = "uc_kep_obs/pays.txt"

year_slice = 2016
column_to_slice = "BDate"
num_of_rows = 1000
dates_rows = [11,12]

```

```python
# Program
read_data(file_name,)
df_to_frequent_items = read_data(csv_file=file_name,
                                 number_rows_to_read = num_of_rows,
                                 slice_year=year_slice,
                                 slice_year_column=column_to_slice,
                                 columns_with_dates=dates_rows)


```
